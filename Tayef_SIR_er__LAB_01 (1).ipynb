Python Lists

```python
x = [1, 5, 4, 6, 1]  # 1D list
print(x)  # Output: [1, 5, 4, 6, 1]

x = [[1, 2], [3, 4]]  # 2D list
print(x)  # Output: [[1, 2], [3, 4]]

print(type(x))  # Output: <class 'list'>
```

Numpy Array in Python

```python
x = np.array([1, 5, 4, 6, 1]) #1d array
print(x + 10)  # Output: [11 15 14 16 11]
print()

x = np.array([[1, 2], [3, 4]]) #2d array
print("2D array : " + str(x))
print()

x = np.array([[[55, 24], [33, 42]],    # 3D NumPy array
              [[12, 22], [32, 42]]])
print(x)
```

Array indexing in numpy

```python
x = np.array([1, 5, 4, 6, 10])
print(x[1:3])
```

```python
(x[1:])
```

```python
(x[-2])
```

```python
(x[:-1])
```

**Numpy Array Slicing**


```python
print(x[0:5:1])
```

```python
print(x[0:5:2])
```

```python
print(x[::2])
```

```python
print(x[::-1])
```

```python
y = np.array([[1, 2, 3, 4, 5],
              [0, 1, 3, 2, 4],
              [2, 3, 4, 6, 7]])
print(y)
```

```python
print(y[1:, 1:3])
```

```python
print(y[:, ::2])
```

```python
I2 = io.imread("sample.jpg")
print(I2.shape)
```

```python
plt.rcParams["figure.figsize"] = (3, 3)
io.imshow(I2)
plt.axis("off")
```

**Display RGB channels separately**

```python
# Display RGB channels separately
plt.rcParams["figure.figsize"] = (15, 5)

plt.subplot(1, 3, 1)
io.imshow(I2[:, :, 0])  # Red channel
plt.title("Red")
plt.axis("off")

plt.subplot(1, 3, 2)
io.imshow(I2[:, :, 1])  # Green channel
plt.title("Green")
plt.axis("off")

plt.subplot(1, 3, 3)
io.imshow(I2[:, :, 2])  # Blue channel
plt.title("Blue")
plt.axis("off")

```
